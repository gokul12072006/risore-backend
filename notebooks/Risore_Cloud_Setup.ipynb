# Risore 1.0 - Free Cloud Setup (Google Colab / Kaggle)

This notebook allows you to run Risore 1.0 entirely for free on Google Colab while using **Google Drive** as the permanent storage for your AI's knowledge base.

```python
# 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 2. Install Dependencies
import os
os.system("pip install -q langchain langchain-community langchain-huggingface langchain-chroma chromadb faiss-cpu sentence-transformers PyPDF2 pdfplumber python-docx markdown gradio pydantic")

# Install Ollama for Linux
import os
os.system("curl -fsSL https://ollama.com/install.sh | sh")
```

```python
# 3. Start Ollama Server in Background
import subprocess
import time

# Start Ollama service
subprocess.Popen(["ollama", "serve"])
time.sleep(3) # Wait for it to start

# Download the lightweight model
import os
os.system("ollama pull qwen2.5:3b")
```

```python
# 4. Set Environment Variables & Run App
import os
os.environ["NOVA_STORAGE_PATH"] = "/content/drive/MyDrive/Risore_DB"
os.environ["NOVA_LLM_MODEL"] = "qwen2.5:3b"

# Assuming the project files are uploaded or cloned to Colab
# os.system("python app.py")
```
